**B2029** 大象喝水

`is_integer()`

```python
raw = input().split(" ")
h = int(raw[0]) # 深度
r = int(raw[1]) # 半径

# 20L = 20000cm^3 | 体积 = 3.14 * h * (r**2)

v = 3.14 * h * (r**2)
total = 20000 / v
print(total if total.is_integer() else int(total) + 1)
```

**P1425** 小鱼的游泳时间

```python
raw = input().split(" ")
a = int(raw[0])
b = int(raw[1])
c = int(raw[2])
d = int(raw[3])

# a:b -> c:d

dur = 60 * (c - a) + (d - b)
h = dur // 60
m = dur - 60 * h
print(h, m)
```

**P5707** 上学迟到 ★

```python
raw = input().split(" ")

a = int(raw[0])
b = int(raw[1])

t = a / b + 10

ct = int(1440 - t)

h = ct // 60
m = ct - h * 60

h -= 16
if h < 0: h += 24

print("%02d:%02d" % (h, m))
```
